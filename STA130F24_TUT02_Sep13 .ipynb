**1.1. Data types: Tuples, Lists, and Dictionaries**
- Tuples: `tuple()`

    - Definition: An ordered, **immutable** collection of elements.
    - Syntax: my_tuple = (1, 2, 3)
    - Usage: Useful when you need a fixed set of values that shouldn't change.
    - Key Point: Tuples cannot be modified (no adding/removing elements).

- Lists: `list()`

    - Definition: An ordered, **mutable** collection of elements.
    - Syntax: my_list = [1, 2, 3]
    - Usage: Use lists when you need a flexible sequence of items.
    - Key Point: Lists can be changed (add/remove elements).

- Dictionaries: `dict()`

    - Definition: A collection of **key-value pairs**.
    - Syntax: my_dict = {'key1': 'value1', 'key2': 'value2'}
    - Usage: Use dictionaries to store data that needs to be labeled.
    - Key Point: Keys must be unique and immutable, values can be changed.

```python
# Q: Demonstrate how a tuple works in Python
# Creating a tuple with different data types
my_tuple = (1, "apple", 3.14, True)
print(my_tuple)  # Output: (1, 'apple', 3.14, True)

# Accessing elements
print(my_tuple[0])  # Output: 1
print(my_tuple[1])  # Output: 'apple'

# Negative indexing
print(my_tuple[-1])  # Output: True

# Slicing a tuple
print(my_tuple[1:3])  # Output: ('apple', 3.14)

# Trying to change an element (will raise an error)
# my_tuple[0] = 10  # Uncommenting this line will raise a TypeError

# Packing
packed_tuple = (10, 20, 30)

# Unpacking
a, b, c = packed_tuple
print(a)  # Output: 10
print(b)  # Output: 20
print(c)  # Output: 30

# Counting the occurrences of a value
print(my_tuple.count(1))  # Output: 2 (counts True as 1)

# Finding the index of an element
print(my_tuple.index("apple"))  # Output: 1

# Nested tuples
nested_tuple = (1, (2, 3), (4, 5, 6))
print(nested_tuple[1])  # Output: (2, 3)
```

```python
# Q: Demonstrate how a list works in Python.
# 1. Creating a List
# A list can contain integers, strings, floats, and even other lists.
my_list = [10, "apple", 3.14, True]
print(my_list)  # Output: [10, 'apple', 3.14, True]

# 2. Accessing Elements
# Lists are indexed starting at 0. You can access elements using their index.
print(my_list[0])  # Output: 10
print(my_list[1])  # Output: 'apple'

# Negative indexing allows you to access elements from the end of the list.
print(my_list[-1])  # Output: True

# 3. Modifying Elements
# Since lists are mutable, you can change their elements.
my_list[0] = 20
print(my_list)  # Output: [20, 'apple', 3.14, True]

# 4. Adding Elements
# You can add elements using append(), insert(), or extend().
my_list.append("banana")  # Adds 'banana' at the end
print(my_list)  # Output: [20, 'apple', 3.14, True, 'banana']

# Using insert() to add an element at a specific position
my_list.insert(1, "orange")  # Adds 'orange' at index 1
print(my_list)  # Output: [20, 'orange', 'apple', 3.14, True, 'banana']

# Using extend() to add multiple elements
my_list.extend([1, 2, 3])  # Adds [1, 2, 3] to the end of the list
print(my_list)  # Output: [20, 'orange', 'apple', 3.14, True, 'banana', 1, 2, 3]

# 5. Removing Elements
# You can remove elements using remove(), pop(), or del.
my_list.remove("apple")  # Removes the first occurrence of 'apple'
print(my_list)  # Output: [20, 'orange', 3.14, True, 'banana', 1, 2, 3]

# Using pop() to remove an element by index and return it
popped_item = my_list.pop(0)  # Removes and returns the first element
print(popped_item)  # Output: 20
print(my_list)  # Output: ['orange', 3.14, True, 'banana', 1, 2, 3]

# Using del to remove an element by index
del my_list[1]  # Deletes the element at index 1
print(my_list)  # Output: ['orange', True, 'banana', 1, 2, 3]

# 6. Slicing Lists
# You can slice lists to create sublists.
sliced_list = my_list[1:4]  # Slices from index 1 to 3 (excludes index 4)
print(sliced_list)  # Output: [True, 'banana', 1]

# 7. Iterating Over a List
# You can loop through a list using a for loop.
for item in my_list:
    print(item)

# 8. List Methods
# Lists have many built-in methods like sort(), reverse(), count(), and index().
my_list = [3, 1, 4, 1, 5, 9]
my_list.sort()  # Sorts the list in ascending order
print(my_list)  # Output: [1, 1, 3, 4, 5, 9]

my_list.reverse()  # Reverses the order of the list
print(my_list)  # Output: [9, 5, 4, 3, 1, 1]

# Counting occurrences of an element
print(my_list.count(1))  # Output: 2

# Finding the index of an element
print(my_list.index(4))  # Output: 2

# 9. Nested Lists
# Lists can contain other lists, allowing for multidimensional data structures.
nested_list = [[1, 2, 3], ["a", "b", "c"]]
print(nested_list[0])  # Output: [1, 2, 3]
print(nested_list[1][1])  # Output: 'b'

# 10. List Comprehensions
# List comprehensions provide a concise way to create lists.
squares = [x ** 2 for x in range(5)]
print(squares)  # Output: [0, 1, 4, 9, 16]
```

```python
# Q: Demonstrate how dict() works in Python.
# 1. Creating a Dictionary
# You can create a dictionary using curly braces {} or the dict() function.

# Using curly braces
my_dict = {"name": "Alice", "age": 25, "city": "Toronto"}
print(my_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'Toronto'}

# Using the dict() function
another_dict = dict(name="Bob", age=30, city="Vancouver")
print(another_dict)  # Output: {'name': 'Bob', 'age': 30, 'city': 'Vancouver'}

# 2. Accessing Values
# Access values by referring to their key inside square brackets or using the get() method.
print(my_dict["name"])  # Output: Alice
print(my_dict.get("age"))  # Output: 25

# Using get() with a default value to avoid KeyError if the key doesn't exist
print(my_dict.get("country", "Canada"))  # Output: Canada

# 3. Adding and Modifying Elements
# Add new key-value pairs or modify existing ones by assigning a value to a key.
my_dict["profession"] = "Engineer"  # Adding a new key-value pair
print(my_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'Toronto', 'profession': 'Engineer'}

my_dict["age"] = 26  # Modifying an existing key
print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'Toronto', 'profession': 'Engineer'}

# 4. Removing Elements
# Use pop(), del, or popitem() to remove elements.

# Using pop() to remove a key and return its value
removed_value = my_dict.pop("city")
print(removed_value)  # Output: Toronto
print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'profession': 'Engineer'}

# Using del to remove a key
del my_dict["profession"]
print(my_dict)  # Output: {'name': 'Alice', 'age': 26}

# Using popitem() to remove and return the last inserted key-value pair
my_dict["country"] = "Canada"
last_item = my_dict.popitem()
print(last_item)  # Output: ('country', 'Canada')
print(my_dict)  # Output: {'name': 'Alice', 'age': 26}

# 5. Checking Key Existence
# Use the `in` keyword to check if a key is in the dictionary.
print("name" in my_dict)  # Output: True
print("city" in my_dict)  # Output: False

# 6. Iterating Over a Dictionary
# You can loop through keys, values, or both using keys(), values(), and items().

# Iterating over keys
for key in my_dict:
    print(key)  # Output: name, age

# Iterating over values
for value in my_dict.values():
    print(value)  # Output: Alice, 26

# Iterating over key-value pairs
for key, value in my_dict.items():
    print(f"{key}: {value}")  # Output: name: Alice, age: 26

# 7. Merging Dictionaries
# Use the update() method to merge another dictionary into an existing one.
my_dict.update({"city": "Toronto", "country": "Canada"})
print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'Toronto', 'country': 'Canada'}

# 8. Dictionary Comprehension
# You can create dictionaries using comprehensions for concise and readable code.
squares = {x: x**2 for x in range(5)}
print(squares)  # Output: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}

# 9. Nested Dictionaries
# Dictionaries can contain other dictionaries, allowing for complex data structures.
nested_dict = {
    "person1": {"name": "Alice", "age": 25},
    "person2": {"name": "Bob", "age": 30},
}
print(nested_dict["person1"]["name"])  # Output: Alice

# 10. Clearing and Copying a Dictionary
# Use clear() to empty a dictionary and copy() to create a shallow copy.
my_dict.clear()  # Clears all items from the dictionary
print(my_dict)  # Output: {}

# Copying a dictionary
original_dict = {"a": 1, "b": 2}
copied_dict = original_dict.copy()
print(copied_dict)  # Output: {'a': 1, 'b': 2}
```

```python
# Q: Demonstrate how numpy.array works in Python.
# Importing the numpy library
import numpy as np

# 1. Creating Arrays with numpy.array
# Creating a 1D array (vector)
array_1d = np.array([1, 2, 3, 4, 5])
print("1D Array:", array_1d)  # Output: [1 2 3 4 5]

# Creating a 2D array (matrix)
array_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2D Array:\n", array_2d)
# Output:
# [[1 2 3]
#  [4 5 6]]

# Creating a 3D array
array_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("\n3D Array:\n", array_3d)
# Output:
# [[[1 2]
#   [3 4]]
#
#  [[5 6]
#   [7 8]]]

# 2. Specifying the Data Type
# By default, NumPy determines the data type, but you can specify it using the dtype parameter.
float_array = np.array([1, 2, 3], dtype=float)
print("\nFloat Array:", float_array)  # Output: [1. 2. 3.]

# 3. Accessing Elements
# Access elements using indices, similar to lists.
print("\nAccessing 2D Array Element (0, 1):", array_2d[0, 1])  # Output: 2

# 4. Array Operations
# Arrays allow element-wise operations without loops.
added_array = array_1d + 10
print("\nAdded 10 to each element:", added_array)  # Output: [11 12 13 14 15]

# Multiplying each element by 2
multiplied_array = array_2d * 2
print("\nMultiplied each element by 2:\n", multiplied_array)
# Output:
# [[ 2  4  6]
#  [ 8 10 12]]

# 5. Slicing Arrays
# You can slice arrays to access subsets of data.
print("\nSlicing the first row of the 2D array:", array_2d[0, :])  # Output: [1 2 3]

# Accessing a column
print("Slicing the second column:", array_2d[:, 1])  # Output: [2 5]

# 6. Reshaping Arrays
# Change the shape of an array using the reshape() method.
reshaped = np.reshape(array_1d, (5, 1))
print("\nReshaped 1D Array to 5x1:\n", reshaped)
# Output:
# [[1]
#  [2]
#  [3]
#  [4]
#  [5]]

# 7. Array Attributes
# Exploring useful attributes of numpy arrays
print("\nArray Shape:", array_2d.shape)  # Output: (2, 3) -> 2 rows, 3 columns
print("Array Size:", array_2d.size)    # Output: 6 -> Total number of elements
print("Array Data Type:", array_2d.dtype)  # Output: int64 (or int32 depending on your system)
print("Number of Dimensions:", array_3d.ndim)  # Output: 3

# 8. Creating Arrays with Ranges
# Creating arrays with evenly spaced values using arange() and linspace()
range_array = np.arange(0, 10, 2)  # Start at 0, end before 10, step of 2
print("\nArray Created with arange:", range_array)  # Output: [0 2 4 6 8]

linspace_array = np.linspace(0, 1, 5)  # 5 values between 0 and 1, inclusive
print("Array Created with linspace:", linspace_array)  # Output: [0.   0.25 0.5  0.75 1.  ]

# 9. Mathematical Operations
# Applying mathematical operations to arrays
squared_array = np.square(array_1d)
print("\nSquared Elements:", squared_array)  # Output: [ 1  4  9 16 25]

# 10. Broadcasting
# Demonstrates how NumPy handles operations between arrays of different shapes
array_broadcast = array_1d + np.array([10])
print("\nBroadcasting Array Addition:", array_broadcast)  # Output: [11 12 13 14 15]

# 11. Array Copy vs. View
# Copy creates a new array, while a view is just a reference to the original data.
copy_array = array_1d.copy()
view_array = array_1d.view()
array_1d[0] = 100
print("\nOriginal Array Modified:", array_1d)  # Output: [100   2   3   4   5]
print("Copied Array Remains Unchanged:", copy_array)  # Output: [1 2 3 4 5]
print("View Reflects Changes:", view_array)  # Output: [100   2   3   4   5]
```

```python
# Q: Use if/else statements in Python to create a program that outputs whether a number is odd or even.
# Function to check if a number is odd or even
def check_odd_even(number):
    if number % 2 == 0:
        print(f"{number} is even.")
    else:
        print(f"{number} is odd.")

# Example usage
number = int(input("Enter a number: "))
check_odd_even(number)
```

```python
# Q: Write Python code that solves the FizzBuzz test.
def fizz_buzz(n):
    for i in range(1, n + 1):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)

# Example usage
n = int(input("Enter the number up to which FizzBuzz should be printed: "))
fizz_buzz(n)
```

## 2. [Monty Hall problem](https://github.com/pointOfive/stat130chat130/blob/main/CHATLOG/wk2/GPT/SLS/00001_gpt3p5_MonteHall_ProblemExplanation_v1.md)
- A game show scenario where you choose between three doors; one hides a prize, and the host reveals a non-prize door after your initial choice. You can then choose to switch or stay.

### Q: Which of the coding structures above you recognize (or do not see) in the Monty Hall simulation code below?

```python
# Monte Hall Simulation Code -- not the only way to code this, but it's what Prof. Schwartz came up with...
# Q: Explain what this Python code is doing.
import numpy as np
all_door_options = (1,2,3)  # tuple
my_door_choice = 1  # 1,2,3
i_won = 0
reps = 100000
for i in range(reps):
    secret_winning_door = np.random.choice(all_door_options)
    all_door_options_list = list(all_door_options)
    # take the secret_winning_door, so we don't show it as a "goat" losing door
    all_door_options_list.remove(secret_winning_door)
    try:
        # if my_door_choice was secret_winning_door then it's already removed
        all_door_options_list.remove(my_door_choice)
    except:
        pass
    # show a "goat" losing door and remove it
    goat_door_reveal = np.random.choice(all_door_options_list)
    all_door_options_list.remove(goat_door_reveal)

    # put the secret_winning_door back in if it wasn't our choice
    # we previously removed it, so it would be shown as a  "goat" losing door
    if secret_winning_door != my_door_choice:
        all_door_options_list.append(secret_winning_door)
    # if secret_winning_door was our choice then all that's left in the list is a "goat" losing door
    # if secret_winning_door wasn't our choice then it's all that will be left in the list

    # swap strategy
    my_door_choice = all_door_options_list[0]

    if my_door_choice == secret_winning_door:
        i_won += 1

i_won/reps
```

```python
# Q: Can you refine the code to be shorter or more efficient?
import numpy as np

all_door_options = [1, 2, 3]  # list (no need for tuple conversion later)
my_door_choice = 1  # 1, 2, or 3
i_won = 0
reps = 100000

for i in range(reps):
    secret_winning_door = np.random.choice(all_door_options)

    # Monty opens a goat door that is neither the chosen door nor the winning door
    remaining_doors = [door for door in all_door_options if door != my_door_choice and door != secret_winning_door]
    goat_door_reveal = np.random.choice(remaining_doors)

    # The player switches to the remaining door (the one not chosen or revealed)
    my_door_choice = [door for door in all_door_options if door != my_door_choice and door != goat_door_reveal][0]

    # Check if the new choice is the winning door
    if my_door_choice == secret_winning_door:
        i_won += 1

# Output the probability of winning by switching
print(i_won / reps)
```
