#### Q: What is the difference between sensitivity and specificity, and why do they matter in medical diagnoses?

<details class="details-example"><summary><u><span style="color:blue">Further detailed</span></u></summary>

- Sensitivity measures how well the model identifies actual positives. Specificity measures how well the model identifies actual negatives.
- In this context, sensitivity measures the ability to correctly identify those with the disease (true positives), while specificity measures the ability to correctly identify those without the disease (true negatives). In medical diagnoses, a high sensitivity helps ensure that few cases are missed, and a high specificity helps avoid false alarms.
</details>

#### Q: What's the difference between **out of sample** versus **in sample predictive performance** and why is it important?
<details class="details-example"><summary><u><span style="color:blue">Further detailed</span></u></summary>

- In-sample performance refers to how well the model predicts data it was trained on, while out-of-sample performance reflects how well the model does on unseen data.
- In-sample performance may overestimate how good the model is since it's tested on data it has already seen. Out-of-sample performance provides a more realistic picture of how the model will perform in the real world.
</details>

## üöß üèóÔ∏è Demo: Lyme Disease and Alternative Metrics [25 minutes]
<!-- *Biases and Unfairness*? <br>Clinical Trials and the 'Adult Income' Dataset -->


> Lyme disease is a tick-borne illness that can cause severe health problems if not treated early. Early symptoms are often nonspecific and can mimic less severe viral illnesses, making accurate diagnosis necessary. A healthtech company has developed a predictive machine learning model intended to assist doctors in diagnosing Lyme disease based on symptoms, history of tick exposure, and initial blood tests. The model was tested in a clinical trial involving patients suspected of having Lyme disease based on their symptoms and exposure history. 




### A. *Confusion matrix* and the standard *metrics* 

|                 | Predict Disease | Predict No Disease |
|-----------------|-----------------|--------------------|
| **Disease**     | 23 (TP)         | 7 (FN)             |
| **No Disease**  | 90 (FP)         | 280 (TN)           |

- **True Positives (TP)**: 23 - The model correctly predicted the disease.
- **False Negatives (FN)**: 7 - The model incorrectly predicted no disease when the disease was present.
- **False Positives (FP)**: 90 - The model incorrectly predicted the disease when it was absent.
- **True Negatives (TN)**: 280 - The model correctly predicted no disease.

**Accuracy**: $\frac{TP + TN}{TP + FN + FP + TN} = \frac{23 + 280}{23 + 7 + 90 + 280} = 0.759$

**Specificity**: $\frac{TN}{FP + TN} = \frac{280}{90 + 280} = 0.757$

**Sensitivity** : $\frac{TP}{TP + FN} = \frac{23}{23 + 7} = 0.767$


#### What‚Äôs the impact of false positives and false negatives in the context of Lyme disease?
<details class="details-example"><summary><u><span style="color:blue">Further detailed</span></u></summary>

- Many false positives (i.e. low specificity) might lead to unnecessary treatment (with potential side effects), while many false negatives (i.e. low sensitivity) could delay crucial treatment, potentially leading to severe health complications. 

### B. Alternative *metrics* that might give a clearer and more direct picture of what's happening

The **metrics** above measure *correctness*, but perhaps we should first consider **metrics** that are instead focused on *incorrectness*. 

**False Positive Rate (1 minus specificity)**: $1 - \frac{TN}{FP + TN} = 1 - \frac{280}{90 + 280} = 1 - 0.757 = 0.243$

Additionally, all of the above **metrics** are measured with respect to the "truth" *which we don't know in practice (and only know for this observed data)*. So we could also instead consider **metrics** which measure performance relative to the prediction that we make by considering a "different denominator".

**Precision**: $\frac{TP}{TP + FP} = \frac{23}{23 +90} = 0.204$

- measure how many of the positive predictions were actually correct

#### Q: What is the ethical concern if the model has a low precision?
<details class="details-example"><summary><u><span style="color:blue">Further detailed</span></u></summary>

- A low precision (i.e. lots of false positives) means many of the positive predictions are wrong, leading to overdiagnosis and unnecessary treatments, which could have side effects and cost implications.
</details>
    
#### Q: Why considering false positive rate?
<details class="details-example"><summary><u><span style="color:blue">Further detailed</span></u></summary>

- Focusing on false positives emphasizes the mistakes made when predicting disease. In a medical context, reducing false positives can avoid harm to patients who do not need treatment.
</details>

<br>
<details class="details-example"><summary><u><span style="color:blue">Key takeaway</span></u></summary>
- In a medical context, reducing false positives can avoid harm to patients who do not need treatment. So we want low false positive rate but high precision.

### C. What does *Class Imbalance* do here?

There are 30 cases of Lyme Disease and 370 individuals that do not have Lyme Disease.
Since only 7.5% of the people here have Lyme Disease what if we just predict no Lyme Disease for everyone? 

|                 | Predict Disease | Predict No Disease |
|-----------------|-----------------|--------------------|
| **Disease**     | 0 (TP)          | 30 (FN)            |
| **No Disease**  | 0 (FP)          | 370 (TN)           |

**Accuracy**: $\frac{TP + TN}{TP + FN + FP + TN} = \frac{0 + 370}{0 + 30 + 0 + 370} = 0.90$

Compare this against the original **metrics** and consider how **Class Imbalance** seems to warp our view of performance.  

**Sensitivity** : $\frac{TP}{TP + FN} = \frac{23}{23 + 7} = 0.767$

**Precision**: $\frac{TP}{TP + FP} = \frac{23}{23 +90} = 0.204$


#### Q: If we focus only on accuracy, what ethical issues might arise when diagnosing Lyme disease?
<details class="details-example"><summary><u><span style="color:blue">Further detailed</span></u></summary>

- Accuracy alone can be misleading, especially with class imbalance. A model that predicts "no disease" for everyone could still have high accuracy, but this would fail to diagnose any actual cases, leading to severe consequences for patients.
    
</details>

#### Q: What happens if we predict no Lyme disease for everyone?
<details class="details-example"><summary><u><span style="color:blue">Further detailed</span></u></summary>

- Sensitivity: $\frac{TP}{TP + FN} = \frac{0}{0 + 30} = 0$
- Precision: Undefined (since no positive predictions).
- Predicting no disease gives a high accuracy but completely misses all true cases of disease, which is dangerous in a medical context. This shows how class imbalance can distort our view of model performance if we only look at accuracy.
    
</details>

### Trade-offs between sensitivity (catching all disease cases) and specificity (avoiding overdiagnosis).

#### Q: In the case of Lyme disease, should we prioritize sensitivity or specificity, and why?
<details class="details-example"><summary><u><span style="color:blue">Further detailed</span></u></summary>

- In many medical contexts, sensitivity is more important because missing a true case of disease (false negative) could have severe consequences. However, specificity is still crucial to avoid unnecessary treatments.
</details>
    
#### Q: In the context of spam email detection, if your priority is to not miss any important emails, should we prioritize sensitivity or specificity, and why?
<details class="details-example"><summary><u><span style="color:blue">Further detailed</span></u></summary>

- Specificity ensures that the filter correctly identifies legitimate emails and keeps them out of the spam folder. This minimizes the risk of important emails being mistakenly marked as spam (false positives).
- By prioritizing specificity, you'll have fewer false positives, meaning more legitimate emails will land in your inbox, and you'll be less likely to miss something important.
</details>
